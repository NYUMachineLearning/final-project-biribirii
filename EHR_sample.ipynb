```python
import random
import os
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
raw_data = '/gpfs/scratch/bcc276/EHR_Dream/train/'
measurement_data_file = 'measurement.csv'
observation_data_file = 'observation.csv'
person_data_file = 'person.csv'
visitoccurrence_data_file = 'visit_occurrence.csv'
drugexposure_data_file = 'drug_exposure.csv'
procedure_data_file = 'procedure_occurrence.csv'
condition_data_file = 'condition_occurrence.csv'
death_data_file = 'death.csv'
```

```python
person = pd.read_csv(
    '{}{}'.format(raw_data, person_data_file),
    index_col=False
)

person['race_concept_id'].astype(float)
person.head()
```

```python
person.drop(['race_source_concept_id', 
             'gender_source_concept_id', 
             'ethnicity_source_concept_id', 
             'provider_id', 'care_site_id', 
             'location_id', 
             'person_source_value', 
             'gender_source_value', 
             'ethnicity_source_value', 
             'race_source_value',
             'year_of_birth',
             'month_of_birth',
             'day_of_birth'
            ], 
            axis=1, 
            inplace = True)
person.head()
```

```python
person['race_concept_id'].replace(8557, 'Native Hawaiian or Other Pacific Islander',inplace=True)
person['race_concept_id'].replace(8657, 'American Indian or Alaska Native',inplace=True)
person['race_concept_id'].replace(8515, 'Asian',inplace=True)
person['race_concept_id'].replace(8516, 'African American',inplace=True)
person['race_concept_id'].replace(8527, 'White',inplace=True)
person['race_concept_id'].replace(8552, 'Unknown',inplace=True)

person['ethnicity_concept_id'].replace(38003564, 'Not Hispanic or Latino',inplace=True)
person['ethnicity_concept_id'].replace(38003563, 'Hispanic or Latino',inplace=True)

person['gender_concept_id'].replace(8532, 'Female',inplace=True)
person['gender_concept_id'].replace(8507, 'Male',inplace=True)

person = person.fillna('NA')
person.head()
```

```python
visit = pd.read_csv(
    '{}{}'.format(raw_data, visitoccurrence_data_file),
    index_col=False
)

visit.head()
```

```python
visit.drop(['discharge_to_source_value', 
            'admitting_source_value', 
            'visit_source_concept_id', 
            'admitting_source_concept_id', 
            'visit_start_datetime', 
            'visit_end_datetime', 
            'provider_id', 
            'care_site_id', 
            'discharge_to_concept_id', 
            'visit_source_value', 
            'visit_type_concept_id',
            'visit_concept_id',
            'preceding_visit_occurrence_id'],
          axis=1,
          inplace=True)
visit.head()
```

```python
death = pd.read_csv(
    '{}{}'.format(raw_data, death_data_file),
    index_col=False
)

death.head()
```

```python
death.drop(['cause_concept_id', 
            'cause_source_value', 
            'death_type_concept_id',
            'death_datetime',
            'cause_source_concept_id'],
          axis=1,
          inplace=True)
death.head()
```

```python
# merge death and visit df by person_id
death_visit = pd.merge(visit, death, how='inner', on=['person_id'])
death_visit.head()
```

```python
# subtract death date from visit_end_date
from datetime import datetime
date_format = "%Y-%m-%d"
date_days = []

for i in range(len(death_visit['visit_end_date'])):
    a = datetime.strptime(death_visit['visit_end_date'][i], date_format)
    b = datetime.strptime(death_visit['death_date'][i], date_format)
    delta = b - a
    # append all dates as days to list
    date_days.append(delta.days)
```

```python
# convert date_days list to Series
date_days = pd.Series(date_days)
date_days.name = 'days_to_death'

# merge date_days with death_visit df
death_visit = death_visit.merge(date_days.to_frame(), left_index=True, right_index=True)
death_visit.head()
```

```python
# condition to check for patients who died within 6 months of visit_end_date
patients_sixmonths = death_visit[death_visit['days_to_death'] <= 183]
patients_sixmonths['person_id'].unique()
```

```python
pt_sixmonths = pd.DataFrame(patients_sixmonths['person_id'].unique())
pt_sixmonths.columns = ['person_id']
pt_sixmonths.head()
```

```python
sample = person.sample(n=5000, replace=False, random_state=1)
sample.head()
```

```python
sample.shape
```

```python
##check to see how many of the deceased within 6 mons patients are present in the sample
sample['person_id'].isin(pt_sixmonths['person_id']).value_counts()
```

```python
##inner merge of the dataframes pulls out only the rows of pts who are present in both - this includes all
##patients who died
death_sample = pd.merge(death, person, how = "inner", on=['person_id'])
```

```python
##merging again with the dataframe of filtered deceased within 6 mons patient ids
death_sample = pd.merge(death_sample, pt_sixmonths, how = "inner", on=['person_id'])
##adding a marker variable for our outcome of interest
death_sample['deceased'] = 1
```

```python
death_sample.head()
```

```python
##adding sampled 5,000 + the deceased patients
full_sample = pd.concat([sample, death_sample], sort=False)
##removing the 22 overlapping records
full_sample = full_sample.drop_duplicates(subset='person_id')
full_sample.head()
```

```python
full_sample.shape
```

```python
##isolating person_ids to pull out relevant details from each dataframe
full_sample_ids = full_sample.drop(labels=['birth_datetime',
                                           'ethnicity_concept_id',
                                           'race_concept_id',
                                           'gender_concept_id',
                                           'death_date',
                                           'deceased'], 
                                   axis=1, 
                                   inplace=False
                                  )
full_sample_ids.head()
```

```python
##creating a dataframe with visit for selected sample
visit_sample = pd.merge(full_sample_ids, visit, how='inner', on=['person_id'])
visit_sample.head()
```

```python
visit_sample.shape
```
