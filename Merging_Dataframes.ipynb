```python
import numpy as np
import pandas as pd
```

```python
labs_data_file = 'labs_sample.csv' # get average and set up tuple or dictionary to call values
person_data_file = 'full_sample.csv' # males = 0, females = 1
visit_data_file = 'visit_sample.csv' # number of visits the patient had within the last year of their last visit
drug_data_file = 'drug_sample.csv' # did the patient take the drug? this will be summed
procedure_data_file = 'procedure_sample.csv' # did the patient get this procedure? this will be summed
condition_data_file = 'condition_sample.csv' # does the patient have this condition? we will drop duplicates for this
```

```python
person = pd.read_csv((person_data_file),index_col=False)
person.drop(person.columns[0], axis=1, inplace=True)
person.head()
```

```python
person['ethnicity_concept_id'].replace('Not Hispanic or Latino', 0, inplace=True)
person['ethnicity_concept_id'].replace('Hispanic or Latino', 1, inplace=True)

person['gender_concept_id'].replace('Male', 0, inplace=True)
person['gender_concept_id'].replace('Female', 1, inplace=True)

person.head()
```

```python
add = pd.get_dummies(person['race_concept_id'])
person = person.join(add)
```

```python
person.drop(['race_concept_id'],axis=1,inplace=True)
person.head()
```

```python
condition = pd.read_csv((condition_data_file),index_col=False)
condition.drop(condition.columns[0], axis=1, inplace=True)
condition.head()
```

```python
add = pd.get_dummies(condition['condition_concept_id'])
condition = condition.join(add)
```

```python
condition.drop(['condition_concept_id','condition_start_date'],
          axis=1,
          inplace=True)
condition.head()
```

```python
condition = condition.groupby('person_id').sum()
condition.head()
```

```python
person_condition = pd.merge(person, condition, how='left', on=['person_id'])
person_condition.head()
```

```python
drugs = pd.read_csv((drug_data_file),index_col=False)
drugs.drop(drugs.columns[0], axis=1, inplace=True)
drugs.head()
```

```python
drugs.drop(['days_supply','verbatim_end_date','quantity','drug_exposure_start_date'],
          axis=1,
          inplace=True)
drugs.head()
```

```python
add = pd.get_dummies(drugs['drug_concept_id'])
drugs = drugs.join(add)
```

```python
#condition['person_id'].astype(str)
drugs = drugs.groupby('person_id').sum()
#condition = condition.groupby('person_id').agg(''.join)
drugs.head()
```

```python
drugs.drop(drugs.columns[1], axis=1, inplace=True)
person_condition_drugs = pd.merge(person_condition, drugs, how='left', on=['person_id'])
person_condition_drugs.head()
```

```python
visits = pd.read_csv((visit_data_file),index_col=False)
visits.drop(visits.columns[0], axis=1, inplace=True)
visits.head()
```

```python
last_visit = visits.sort_values(by=['visit_end_date']).drop_duplicates(subset='person_id', keep='last')
last_visit.head()
```

```python
last_visit.drop(['visit_start_date','visit_occurrence_id'],
          axis=1,
          inplace=True)
last_visit.head()
```

```python
last_visit.rename(columns = {'visit_end_date':'last_visit'}, inplace = True)
last_visit.head()
```

```python
visit_number = pd.merge(visits, last_visit, how='inner', on=['person_id'])
visit_number.head()
```

```python
from datetime import datetime
date_format = "%Y-%m-%d"
date_days = []
for i in range(len(visit_number['visit_end_date'])):
    a = datetime.strptime(visit_number['visit_end_date'][i], date_format)
    b = datetime.strptime(visit_number['last_visit'][i], date_format)
    delta = b - a
    # append all dates as days to list
    date_days.append(delta.days)
print(date_days)
# convert date_days list to Series
date_days = pd.Series(date_days)
date_days.name = 'days_to_last_visit'
# merge date_days with death_visit df
visit_number = visit_number.merge(date_days.to_frame(), left_index=True, right_index=True)
```

```python
# condition to check for patients who died within 6 months of visit_end_date
patients_visits = visit_number[visit_number['days_to_last_visit'] <= 365]
patients_visits['visits_last_year'] = 1
patients_visits.head()
```

```python
patients_visits.drop(['visit_start_date','visit_occurrence_id','days_to_last_visit','visit_end_date','last_visit'],
          axis=1,
          inplace=True)
patients_visits.head()
```

```python
patients_visits = patients_visits.groupby('person_id').sum()
patients_visits.head()
```

```python
person_condition_drugs_visits = pd.merge(person_condition_drugs, patients_visits, how='left', on=['person_id'])
person_condition_drugs_visits.head()
```

```python
procedure = pd.read_csv((procedure_data_file),index_col=False)
procedure.drop(procedure.columns[0], axis=1, inplace=True)
procedure.head()
```

```python
procedure.drop(['procedure_date','visit_occurrence_id'],
          axis=1,
          inplace=True)
procedure.head()
```

```python
add = pd.get_dummies(procedure['procedure_concept_id'])
procedure = procedure.join(add)
procedure.drop(['procedure_concept_id'],axis=1,inplace=True)
procedure.head()
```

```python
procedure.drop([0],axis=1,inplace=True)
procedure = procedure.groupby('person_id').sum()
procedure.head()
```

```python
person_condition_drugs_visits_procedures = pd.merge(person_condition_drugs_visits, procedure, how='left', on=['person_id'])
person_condition_drugs_visits_procedures.head()
```

```python
person_age = pd.merge(person,visit_number,how='inner',on=['person_id'])
person_age.head()
```

```python
person_age.drop(['ethnicity_concept_id','African American','Native Hawaiian or Other Pacific Islander','White','gender_concept_id','deceased','visit_occurrence_id',
                'visit_end_date','visit_start_date','days_to_last_visit'],axis=1,inplace=True)
person_age.head()
```

```python
date_format = "%Y-%m-%d"
date_days = []
for i in range(len(person_age['last_visit'])):
    a = datetime.strptime(person_age['birth_datetime'][i], date_format)
    b = datetime.strptime(person_age['last_visit'][i], date_format)
    delta = b - a
    # append all dates as days to list
    date_days.append(delta.days)
print(date_days)
# convert days to years
date_years = [i / 365 for i in date_days]
# convert date_days list to Series
date_years = pd.Series(date_years)
date_years.name = 'age'
# merge date_days with death_visit df
person_age = person_age.merge(date_years.to_frame(), left_index=True, right_index=True)
person_age.head()
```

```python
person_age.drop(['birth_datetime','last_visit'],axis=1,inplace=True)
person_age.head()
```

```python
person_age.drop_duplicates(inplace=True)
person_age.head()
```

```python
df = pd.merge(person_condition_drugs_visits_procedures,person_age,how='left',on=['person_id'])
df.head()
```

```python
df.drop(['birth_datetime'],axis=1,inplace=True)
```

```python
labs = pd.read_csv((labs_data_file),index_col=False)
labs = labs[['person_id', 
             'measurement_concept_id', 
             'value_as_number']]
labs.head()
```

```python
# take average of lab values
# labs.groupby(['person_id', 'measurement_concept_id']).agg([np.average])
# alternative way
labs_average = labs.groupby(['person_id', 'measurement_concept_id'])['value_as_number'].mean().reset_index()
```

```python
# pivot the df
labs_average = labs_average.pivot(index='person_id', columns='measurement_concept_id', values='value_as_number')
labs_average.head()
```

```python
df = pd.merge(df, labs_average, how='left', on=['person_id'])
df.head()
```

```python
# drop all columns that have >= 95% NA
df = df.loc[:, df.isnull().mean() < 0.95]
df.head()
```
